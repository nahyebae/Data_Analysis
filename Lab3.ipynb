```python
import folium
import pandas as pd
```

```python
data = pd.read_csv("택시승차대 현황.csv", encoding='cp949')
data
```

```python
m = folium.Map(location = [37.54539703, 126.9525641], zoom_start = 15)
taxi = data.loc[:, ["위치명", "위도", "경도"]]
for index, row in taxi.iterrows():
    folium.Marker([row["위도"], row["경도"]], tooltip = row["위치명"], 
                  icon = folium.Icon(color = 'black', icon = 'fa-solid fa-taxi',
                                     prefix = "fa-solid")).add_to(m)
    folium.Circle([row["위도"], row["경도"]],
                        radius = 100,
                        color = '#000000',
                        fill_color = '#ff0000',
                        popup = "Circle").add_to(m)
m
```
