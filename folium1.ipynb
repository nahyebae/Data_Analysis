```python
import folium
```

```python
m = folium.Map(location = [37.54425916885864, 127.03759056993367], zoom_start = 15)
m
```

```python
m = folium.Map(location = [37.54425916885864, 127.03759056993367], zoom_start = 30, 
               tiles = "openstreetmap", width = 800, height = 600)
m
```

```python
folium.Marker([37.54425916885864, 127.03759056993367], popup = "팝업", tooltip = "툴팁").add_to(m)
m
```

```python
folium.Marker([37.54425916885864, 127.03759056993367], popup = "팝업", tooltip = "툴팁",
              icon = folium.Icon(color = 'green', icon = 'fa-solid fa-tree',
                                 prefix = "fa-solid")).add_to(m)
m
```

```python
m.add_child(folium.ClickForMarker(popup="클릭"))
m.add_child(folium.LatLngPopup())
m
```

```python
folium.CircleMarker([37.54425916885864, 127.03759056993367],
                    radius = 100,
                    color = '#0000ff', # 원 윤곽선 색
                    fill_color = '#c694eb', # 원 배경 색
                    popup = "CircleMaker").add_to(m)
folium.Circle([37.54425916885864, 127.03759056993367],
                    radius = 100,
                    color = '#ff0000',
                    fill_color = '#ffffff',
                    popup = "Circle").add_to(m)
m
```
