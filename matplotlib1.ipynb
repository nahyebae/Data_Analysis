```python
import numpy as np
x = [100,400,900]
y = np.arange(3)
plt.barh(y, x, color = ['r','g','b'], height = 0.7)
# plt.plot(x,y)
plt.yticks(y, ['2018','2019','2020'])
# plt.yticks(x, ['a','b','c'])
plt.show()
```

```python
ratio = [24, 28,12, 36]
labels = ['A', 'B', 'C', 'D'] # 반시계 방향
wedgepros = {'width': 0.7, 'edgecolor': 'k'}

plt.pie(ratio, labels = labels, wedgeprops=wedgepros, autopct = '%.1f%%', explode=(0,0.1,0.1,0.1), colors=['pink','skyblue','red','gray'], counterclock = False)
```

```python
# 실습 1. 꺽은선 그래프 그리기
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales_2019 = [100,120,140,110,130,150,160,170,180,200,190,210]
sales_2020 = [90,110,130,120,140,160,170,160,150,180,200,190]

plt.title("Monthly Sales Comparison (2019-2020)")
plt.plot(months,sales_2019, label='2019')
plt.plot(months,sales_2020, label='2020')
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

```python
# 실습 2. 막대 그래프 그리기
categories = ['Category 1','Category 2','Category 3','Category 4','Category 5']
data = [20, 35, 15, 27, 45]

plt.title('Bar Chart')
plt.xlabel('Categories')
plt.xticks(rotation=45)
plt.ylabel('Values')
plt.bar(categories, data)
plt.ylim(0,50) # y축 최대값
plt.grid()
plt.show()
```

```python
# 실습 3. 파이차트 그리기
labels = ['Apple', 'Banana','Mango', 'Blueberry']
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)
wedgepros = ['e']
plt.pie(sizes, labels = labels, explode = explode, autopct='%.1f%%', startangle = 90, shadow = True)
plt.show()
```

```python
# 실습 4. sin & cos 함수 그리기
x = np.linspace(0, 10, 100)

# sin과 cos 함수의 값을 계산
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.title('Sin and Cosine Functions')
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(x, y_sin, 'b', label = 'sin(x)')
plt.plot(x, y_cos, 'r', label = 'cos(x)')
plt.grid()
plt.legend()
plt.show()
```
